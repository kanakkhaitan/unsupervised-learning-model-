```python
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_style('darkgrid')
%matplotlib inline
```

```python
iris_df=sns.load_dataset('iris')
iris_df
```

```python
sns.scatterplot(data=iris_df,x='sepal_length',y='petal_length',hue='species')
```

```python
numeric_cols=["sepal_length","sepal_width","petal_length","petal_width"]
X=iris_df[numeric_cols]
X
```

```python
from sklearn.cluster import KMeans
model=KMeans(n_clusters=3,random_state=42)
model.fit(X)
```

```python
model.cluster_centers_
```

```python
X
```

```python
preds=model.predict(X)
preds
```

```python
X['preds']=preds
sns.scatterplot(data=X,x='sepal_length',y='petal_length',hue='preds')
centers_x,centers_y=model.cluster_centers_[:,0],model.cluster_centers_[:,2]
plt.plot(centers_x,centers_y,'xb')
#x = cross marker
```

```python
X['preds']=preds
X
```

```python
model.inertia_
#Lower inertia means the data points are closer to their assigned cluster centers, so clusters are tighter.

#But lower isn't always better â€” adding more clusters always reduces inertia (even if it's not meaningful), so it's important to balance it with model complexity.

```

```python
model=KMeans(n_clusters=6,random_state=42).fit(X)

preds=model.predict(X)
preds
```

```python
model.cluster_centers_
```

```python
X['preds']=preds
sns.scatterplot(data=X,x='sepal_length',y='petal_length',hue='preds')
```

```python
model.inertia_
```

```python
options=range(2,11)
inertias=[]
for n_clusters in options:
  model=KMeans(n_clusters,random_state=42).fit(X)
  inertias.append(model.inertia_)
plt.title("No of Clusters vs. Inertia")
plt.plot(options,inertias,'-o')
plt.xlabel('No of Clusters(k)')
plt.ylabel('Inertia')
#'-' = solid line
#'o' = circular markers at each data point
#elbow curve
```

```python
#eps (epsilon) is the maximum distance between two points for them to be considered neighbors.
#min_samples is the minimum number of points (including the point itself) required to form a dense region.
from sklearn.cluster import DBSCAN
model=DBSCAN(eps=1.1,min_samples=4).fit(X)
#dir(model) #this line gives all the attributes of the model
```

```python

import numpy as np
print(np.unique(model.labels_))
```

```python
model.core_sample_indices_
```

```python
model.labels_
```

```python
sns.scatterplot(data=X,x='sepal_length',y='petal_length',hue=model.labels_)
#core point,edge point,outlier(nose points)
```
